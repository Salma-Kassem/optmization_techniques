### Load MNIST

```python
from tensorflow.keras.datasets import mnist

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

### Preprocessing

```python
## Reshaping Train Images
x_train = train_images.reshape(train_images.shape[0], (28*28)) / 255.0
x_test = test_images.reshape(test_images.shape[0], (28*28)) / 255.0

## One-hot vectors of train labels
y_train = np.eye(10)[train_labels]
y_test = np.eye(10)[test_labels]
```

### Build Model

```python
def train_mnist(optimizer):
    model = keras.models.Sequential()
    model.add(keras.Input(shape=(784,)))
    model.add(keras.layers.Dense(units=50, activation='relu'))
    model.add(keras.layers.Dense(units=10, activation='sigmoid'))
    model.summary()
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
sgd = keras.optimizers.SGD(learning_rate=0.003)
sgd_momentum = keras.optimizers.SGD(learning_rate=0.003, momentum=0.9, nesterov=True)

model_sgd = train_mnist(sgd)
model_sgd_momentum = train_mnist(sgd_momentum)
```

```python
output_sgd = model_sgd.fit(x_train, y_train, epochs=20, batch_size=32, validation_data=(x_test, y_test)) ## 60000 /32 = 1875
```

```python
output_sgd_momentum = model_sgd_momentum.fit(x_train, y_train, epochs=20, batch_size=32, validation_data=(x_test, y_test)) ## 60000 /32 = 1875
```

```python
model_adagrad = train_mnist("adagrad")
model_rmsprop = train_mnist("rmsprop")
model_adam = train_mnist("adam")
```

```python
output_adagrad = model_adagrad.fit(x_train, y_train, epochs=20, batch_size=32, validation_data=(x_test, y_test)) ## 60000 /32 = 1875
```

```python
output_rmsprop = model_rmsprop.fit(x_train, y_train, epochs=20, batch_size=32, validation_data=(x_test, y_test)) ## 60000 /32 = 1875
```

```python
output_adam = model_adam.fit(x_train, y_train, epochs=20, batch_size=32, validation_data=(x_test, y_test)) ## 60000 /32 = 1875
```

```python
import matplotlib.pyplot as plt

def plot(output):
    train_acc = output.history['accuracy']
    train_loss = output.history['loss']
    val_acc = output.history['val_accuracy']
    val_loss = output.history['val_loss']
    plt.plot(train_acc, label='train_accuracy')
    plt.plot(val_acc, label='val_accuracy')
    plt.legend()
    plt.show()

    plt.plot(train_loss, label='train_loss')
    plt.plot(val_loss, label='val_loss')
    plt.legend()
    plt.show()
```

```python
def plot_optimizers(outputs, labels):
    for output, label in zip(outputs, labels):
        # train_acc = output.history['accuracy']
        # train_loss = output.history['loss']
        # val_acc = output.history['val_accuracy']
        val_loss = output.history['val_loss']
        # plt.plot(train_acc, label='train_accuracy')
        # plt.plot(val_acc, label='val_accuracy')
        # plt.legend()
        # plt.show()

        # plt.plot(train_loss, label='train_loss')
        plt.plot(val_loss, label='val_loss_'+label)
        plt.legend()
    plt.show()
```

```python
plot_optimizers([output_sgd, output_sgd_momentum, output_adagrad, output_rmsprop, output_adam], ["sgd", "sgd_momentum", "adagrad", "rmsprop", "adam"])
```
